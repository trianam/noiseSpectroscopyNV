```python
minMax = (  #from data creation
            (0.0005, 0.01),
            (0.1, 1),
            (50, 1000),
            (0.001, 0.01),
        )

dataset = np.load("data/dataset32-REAL-R1b.npz")
# dataset = np.load("data/dataset32-REAL-R3.1.npz")
# limits = [0, 201]
limits = None

if limits is None:
    coh = dataset['coh']
else:
    coh = dataset['coh'][:,limits[0]:limits[1]]

par = dataset['par']

# for i in range(par.shape[1]):
#     par[:,i] -= minMax[i][0]
#     par[:,i] /= minMax[i][1] - minMax[i][0]
```

```python
def plot(index=0):
    plt.plot(coh[index], label="[{:.2f}, {:.2f}, {:.2f}, {:.2f}]".format(*par[index]))
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', ncol=1)
    plt.xlabel(r'$t_1 \; [\mu{\rm s}]$',fontsize=14)
    plt.ylabel(r'$C(t)$',fontsize=14)
    plt.ylim(bottom=-0.05)
    plt.show()
```

```python
for i in range(len(par)):
    plot(i)
```

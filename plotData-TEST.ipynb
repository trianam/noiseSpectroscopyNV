```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
minMax = (  #from data creation
            (0.0005, 0.01),
            (0.1, 1),
            (50, 1000),
            (0.001, 0.01),
        )

dataset = np.load("data/dataset32.npz")
coh = dataset['coh']
par = dataset['par'][:,:-1]

for i in range(par.shape[1]):
    par[:,i] -= minMax[i][0]
    par[:,i] /= minMax[i][1] - minMax[i][0]
```

```python
par.shape
```

```python
def plotRange(index, base=[0,0,0], jump=1):
    '''
    index: what to explore
    base: the value for the other 3 indexes in order
    jump: plot every x
    '''
    if len(base) == 3:
        base.insert(index, 0)
    elif len(base) == 4:
        base[index] = 0
    else:
        raise Exception("Use 3 or 4 values")
    calcBase = base[3] + base[2]*40 + base[1]*40*40 + base[0]*40*40*40
    for i in range(calcBase, 40**(4-index) + calcBase, jump*40**(4-index-1)):
        plt.plot(coh[i], label="[{:.2f}, {:.2f}, {:.2f}, {:.2f}]".format(*par[i]))
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', ncol=2 if jump<=2 else 1)
    plt.show()
```

```python
plotRange(3)
```

```python
plotRange(2)
```

```python
plotRange(1)
```

```python
plotRange(0)
```

```python
plotRange(3, base=[0,0,20,0])
```

```python
%matplotlib inline
import numpy as np
import h5py
import matplotlib.pyplot as plt
```

```python
dataset = h5py.File("data/dataset32AugV1.hdf5", 'r')
# numPoints =  40
# dataset = h5py.File("data/dataset32AugV2test.hdf5", 'r')
# dataset = h5py.File("data/dataset32AugV3test.hdf5", 'r')
numPoints =  2
numSamples = 10
coh = dataset['coh']
par = dataset['par']
parN = dataset['parNorm']
```

```python
parN.shape
```

```python
def plotValues(values=[0,0,0,0], var=0, indexes=False):
    if len(values) != 4:
        raise Exception("Wrong length")
    if indexes:
        index = values
    else:
        if min(values) < 0 or max(values) > 1:
            raise Exception("Wrong values")
        index = [round(v*(numPoints-1)) for v in values]
    if type(var) is int:
        var = [var]
    for v in var:
        calcBase = v + (index[3] + index[2]*numPoints + index[1]*(numPoints**2) + index[0]*(numPoints**3)) * numSamples
        plt.plot(coh[calcBase], label="[{:.2f}, {:.2f}, {:.2f}, {:.2f}]({})".format(*parN[calcBase],v))
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', ncol=1)
    plt.xlabel(r'$t_1 \; [\mu{\rm s}]$',fontsize=14)
    plt.ylabel(r'$C(t)$',fontsize=14)
    plt.ylim(bottom=-0.05)
    plt.show()
```

```python
plotValues([1,1,1,1], indexes=True)
```

```python
plotValues([1,1,1,1], indexes=True)
```

```python
plotValues([1,1,1,1], indexes=True, var=range(10))
```

```python
plotValues([0.5,0.5,0.5,0.5])
```
